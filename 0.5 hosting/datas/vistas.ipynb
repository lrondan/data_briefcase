```python
import pandas as pd
import matplotlib.pyplot as mpl
import numpy as np
import seaborn as sns
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.model_selection import cross_val_score
```

```python
#data collect
with open('./datalake.csv','r',encoding='utf-8')as file:
    df1 = pd.read_csv(file)
    df1 = pd.DataFrame(data=df1)

x_values =df1[['Dorms','Baths','Built Area','Total Area','Parking']]
y_values =df1[['Price_USD','Price_UF']]

#data train
lr = LinearRegression()
lr.fit(x_values, y_values)
```

```python
Yhat = lr.predict(x_values)
sns.distplot(Yhat, color='g')
sns.distplot(y_values, color='r')
```

```python
#scores
scores = cross_val_score(lr, x_values, y_values, cv=3)
sns.distplot(scores, color='b')
```
