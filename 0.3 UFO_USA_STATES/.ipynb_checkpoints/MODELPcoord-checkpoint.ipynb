```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
import csv
```

```python
with open('./UFOs_coord.csv','r')as file:
    archivo = pd.read_csv(file)
    df1 = pd.DataFrame(data=archivo)
```

```python
lat = df1[['lat']]
lng = df1['lng']
state = df1['State']
Shape = df1['Shape']
```

```python
lm = LinearRegression()
lm.fit(lat,lng)
```

```python
lng_hat = lm.predict(lat)
```

```python
############   MODELO SIMPLE DE LA REGRESION LINEAL, DEMUESTRA LA PROBABILIDAD DE EN QUE COORDENADA DE LONGITUD PUEDE 
############   APARECER UN UFO...
```

```python
axis1 = sns.distplot(lng, hist=False, color='r')
sns.distplot(lng_hat, hist=False, color='g',ax=axis1)
```
