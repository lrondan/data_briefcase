```python
import pandas as pd
import sqlite3
import matplotlib.pyplot as mpl
```

```python
conn = sqlite3.connect('./base_world.sqlite')
df = pd.read_sql('SELECT * FROM actrices', conn)
```

```python
category = df['category'].value_counts()
category_total = category.sum()
category_porcent = (category/category_total)*100
```

```python
category_porcent.plot(kind='pie',
              title='Cantidad de actrices por category',
              colormap='RdBu',
              ylabel='category',
              legend=True)
mpl.show()
```
