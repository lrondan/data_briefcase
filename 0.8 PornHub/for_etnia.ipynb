```python
import pandas as pd
import sqlite3
import matplotlib.pyplot as mpl
```

```python
conn = sqlite3.connect('./base_world.sqlite')
df = pd.read_sql('SELECT * FROM actrices', conn)
```

```python
df_etnia = df['etnia'].value_counts()
df_etnia.plot(kind='barh',
              title='Cantidad de actrices por etnia',
              xlabel='total',
              colormap='RdBu')
```
