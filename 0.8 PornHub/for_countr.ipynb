```python
import pandas as pd
import sqlite3
import matplotlib.pyplot as mpl
```

```python
conn = sqlite3.connect('./base_world.sqlite')
df = pd.read_sql('SELECT * FROM actrices', conn)
```

```python
df_country = df['country'].value_counts().nlargest(10)
df_country.plot(kind='pie',
              title='Cantidad de actrices por paises top 10',
              ylabel='total',
              colormap='RdBu')
```
