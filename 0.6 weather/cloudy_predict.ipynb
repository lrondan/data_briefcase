```python
import numpy as np
from sklearn.linear_model import LinearRegression
import pandas as pd
```

```python
with open('./clima_canada.csv','r')as file1:
    df = pd.read_csv(file1)
    df = pd.DataFrame(data=df)
```

```python
df
```

```python
lr = LinearRegression()
z = df[['Temperatura', 'Humedad', 'Viento', 'Precipitacion(%)']]
y = df['Nublado2']
lr.fit(z,y)
```

```python
###### CODIGO PARA PREDECIR EL ESTADO DEL TIEMPO A PARTIR DE ALGUNAS VARIABLES
```

```python
t = 16 #temperatura
v = 71 #velocidad del viento
h = 9.0 #humedad
p = 80 #precipitacion en %
# nuevo dataframe
d = {'Temperatura':[t],'Humedad':[h], 'Viento':[v],  'Precipitacion(%)':[p]}
x_valors = pd.DataFrame(data=d)
x_valors
```

```python
Yhat = lr.predict(x_valors)
valor_Z_Yhat = int(Yhat)
```

```python
if valor_Z_Yhat == 100:
    print('clear-->100')
elif valor_Z_Yhat ==101:
    print('cloudy-->101')
elif valor_Z_Yhat ==102:
    print('overcast-->102')
else:
    print('partly cloudy-->103')
```
