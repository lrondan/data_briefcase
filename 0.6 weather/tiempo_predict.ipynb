```python
import numpy as np
from sklearn.linear_model import LinearRegression
import pandas as pd
```

```python
with open('./clima_canada.csv','r')as file1:
    df = pd.read_csv(file1)
    df = pd.DataFrame(data=df)
df.head()
```

```python
lr = LinearRegression()
x = df[['Nublado2', 'Presion','UV', 'Season_id','Visibilidad']]
y = df['Tiempo_id']
lr.fit(x, y)
```

```python
# para hacer una nueva prediccion hay q tener en cuenta que se hizon nuevos ajustes para que el algoritmo pueda entender el set de datos
# por ejemplo: 
# NUBLADO se le asigna un numero del 100-103
# Season se le asigna un numero del 1000-1003
# Tiempo se le asigna un numero del 0-3
```

```python
N2 = 102 #nublado
P = 1001.325 #presion
UV = 3 # indice de los rayos UV
V = 4.8 #visibilidad
S_ID = 1001 #Temporada 
nuevos_datos = {'Nublado2':[N2], 'Presion':[P],'UV':[UV], 'Season_id':[S_ID],'Visibilidad':[V]}
x_data = pd.DataFrame(data=nuevos_datos)
x_data
```

```python
Yhat = lr.predict(x_data)
Tiempo_pred = int(Yhat)
```

```python
if Tiempo_pred == 0:
    print('Rainy-->0')
elif Tiempo_pred ==1:
    print('Cloudy-->1')
elif Tiempo_pred==2:
    print('Sunny-->2')
else:
    print('Snowy-->3')
```
