```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import seaborn as sns
import matplotlib.pyplot as mpl
```

```python
with open('./nset.csv', 'r')as file:
    df = pd.read_csv(file)
    df = pd.DataFrame(data=df)
```

```python
lr = LinearRegression()
x = df[['distancia']]
y = df['tiempo']
```

```python
lr.fit(x, y)
```

```python
Yhat = lr.predict(x)
```

```python
sns.distplot(Yhat, color='g')
mpl.title('Densidad de prediccion del tiempo')
mpl.xlabel('escala de t [s]')
```
