```python
#Modelo de prediccion de migracion en canada para el ano proximo por continentes.
```

```python
import pandas as pd
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

```python
with open('./new_train.csv','r')as file:
    archi = pd.read_csv(file)
    df = pd.DataFrame(data=archi)
```

```python
lm = LinearRegression()
```

```python
#seleccion de valores
Cont = df['Etiquetas de fila']
v00 = df[['Suma de 2000','Suma de 2001','Suma de 2002','Suma de 2003','Suma de 2004','Suma de 2005','Suma de 2006','Suma de 2007']]

v_comparable_08 = df[['Suma de 2008','Suma de 2009','Suma de 2010','Suma de 2011','Suma de 2011','Suma de 2012']]
v_comparable_13 = df['Suma de 2013']
```

```python
lm.fit(v00,v_comparable_08)
```

```python
Yhat = lm.predict(v00)
```

```python
ydata = pd.DataFrame(data=Yhat)
predict0 = ydata[0]
predict1 = ydata[1]
predict2 = ydata[2]
predict3 = ydata[3]
predict4 = ydata[4]
predict5 = ydata[5]
media = ydata.median(axis=1)
```

```python
sns.distplot(predict0, color='b')
sns.distplot(predict1, color='b')
sns.distplot(predict2, color='b')
sns.distplot(predict3, color='b')
sns.distplot(predict4, color='b')
sns.distplot(predict5, color='b')
sns.distplot(media, color='r')
```
