```python
import pandas as pd
import matplotlib.pyplot as mpl
import numpy as np
```

```python
with open('./new.csv','r')as file:
    archi = pd.read_csv(file)
    df = pd.DataFrame(data=archi)
newdf = df.drop([6])
newdf = pd.DataFrame(data=newdf)
```

```python
xploteable = newdf.iloc[0:,0]
yploteable = newdf.iloc[0:,1]
```

```python
mpl.bar(xploteable, yploteable)
mpl.ylabel('Total')
mpl.xlabel('Continentes')
mpl.title('Total de inmigrantes')
mpl.savefig('./bar.png')
mpl.show()
```
